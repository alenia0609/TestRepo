```python
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians
from folium.features import DivIcon

# 1. Add MousePosition to map
mouse_position = MousePosition(
    position='topright',
    separator=' | ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat/Lng:'
)
site_map.add_child(mouse_position)

# 2. Distance function (km)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c

# 3. Choose one launch site
launch_site_name = 'CCAFS LC-40'
launch_site = launch_sites_df[launch_sites_df['Launch Site'] == launch_site_name].iloc[0]
launch_coordinate = [launch_site['Lat'], launch_site['Long']]

# Example proximity point (replace with coordinates from MousePosition)
proximity_coordinate = [28.56367, -80.57163]

# Calculate distance
distance_km = calculate_distance(
    launch_coordinate[0], launch_coordinate[1],
    proximity_coordinate[0], proximity_coordinate[1]
)

# Re-create map centered on launch site
site_map = folium.Map(location=launch_coordinate, zoom_start=13)

# Launch site marker
folium.Marker(
    location=launch_coordinate,
    popup=launch_site_name,
    icon=folium.Icon(color='red')
).add_to(site_map)

# Proximity marker
folium.Marker(
    location=proximity_coordinate,
    popup=f'Proximity Point<br>Distance: {distance_km:.2f} km',
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(site_map)

# Draw line between launch site and proximity
folium.PolyLine(
    locations=[launch_coordinate, proximity_coordinate],
    weight=3,
    color='orange'
).add_to(site_map)

# Add distance label
distance_marker = folium.Marker(
    location=[
        (launch_coordinate[0] + proximity_coordinate[0]) / 2,
        (launch_coordinate[1] + proximity_coordinate[1]) / 2
    ],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_km:.2f} KM</b></div>'
    )
)
site_map.add_child(distance_marker)

# Add mouse position again
site_map.add_child(mouse_position)

site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
